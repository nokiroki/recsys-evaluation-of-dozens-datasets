```python
import os

import pandas as pd
```

```python
tvaudience = pd.read_csv(
    os.path.join("..", "..", "data", "tvaudience", "tv-audience-dataset.csv"),
    header=None,
    names=["channelID", "slot", "week", "genreID", "subGenreID", "userID", "programID", "eventID", "duration"],
)
```

```python
#  week from 1 to 19. Weeks 14 and 19 should not be used because they contain errors
tvaudience = tvaudience[(tvaudience.week < 14) | (tvaudience.week > 19)]
```
